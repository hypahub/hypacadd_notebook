# Weighted Margin Classifier (QNN)
*Reformulation of [PennyLane's Margin Classifier](https://pennylane.ai/qml/demos/tutorial_multiclass_classification.html). Disabled random batch sampling from training set per iteration.*
- **Weighted**: weighted loss function to handle class imbalance
- **Margin Classifier**: classifies data over classes using the [margin loss (SVM)](http://cs231n.github.io/linear-classify/)
  - $L(x,y) = \sum\limits_{j≠y}  max(0,s_{j}−s_{y}+\vartriangle)$ where $\vartriangle$ denotes the margin
  - Margin loss requires the score for the correct class is at least larger than the score of incorrect classes by the specified margin 
- **QNN**: layers of qubits with each quantum node representing the classifier for a class 

### Hyperparameters
- **layer = 6**: quicker and more stable convergence (than layer = 3) for both accuracy and cost
- **margin = 0.15**: higher accuracy and lower cost (than margin = 0.5) after 50 iterations

### Input Data Preparation
- Standarized original data
- Performed **PCA** and determined four principal components
- Performed **K-means** on components and assigned data to three clusters
- Stratified data by both *target* and *cluster*
  - Allocated half of each grouping for training and the other half for validation

### Intermediate Results
- For each iteration, we report:
  - The model's *cost* and its current *accuracy* on the train, validation and train + validation (whole) sets
  - TPR, FPR, F1-score, etc.
  - Weighted TP and FN counts accounting for class imbalance

### Saved Outputs
- TSV of prediction and truth values on test set
  - Includes scores given by each classifier used to predict
- Final model saved using pickle
- StandardScaler fitted on the training set saved using pickle
- Cost and accuracy convergence plots for both the train and validation sets

```python
def variational_classifier(q_circuit, params, feat):
    weights = params[0]
    bias = params[1]
    return q_circuit(weights, feat=feat) + bias

def multiclass_svm_loss(q_circuits, all_params, feature_vecs, true_labels, weights, num_classes= None, margin = None):
    loss = 0
    num_samples = len(true_labels)
    
    for i, feature_vec in enumerate(feature_vecs):
        # Compute the score given to this sample by the classifier corresponding to the
        # true label. So for a true label of 1, get the score computed by classifer 1,
        # which distinguishes between "class 1" or "not class 1".
        s_true = variational_classifier(
            q_circuits[int(true_labels[i])],
            (all_params[0][int(true_labels[i])], all_params[1][int(true_labels[i])]),
            feature_vec,
        )
        s_true = s_true.float()
        li = 0

        # Get the scores computed for this sample by the other classifiers
        for j in range(num_classes):
            if j != int(true_labels[i]):
                s_j = variational_classifier(
                    q_circuits[j], (all_params[0][j], all_params[1][j]), feature_vec
                )
                s_j = s_j.float()
                li += torch.max(torch.zeros(1).float(), s_j - s_true + margin)
        loss += li * weights[int(true_labels[i])]

    return loss

def classify(q_circuits, all_params, feature_vecs, labels=None, num_classes = None):
    predicted_labels = []
    predScore = []
    for i, feature_vec in enumerate(feature_vecs):
        scores = np.zeros(num_classes)
        for c in range(num_classes):
            score = variational_classifier(
                q_circuits[c], (all_params[0][c], all_params[1][c]), feature_vec
            )
            scores[c] = float(score)
        predScore.append(scores)
        pred_class = np.argmax(scores)
        predicted_labels.append(pred_class)
    return predicted_labels, predScore
```

```python
def select(data, ind , num = 500):
    rate1 = num/9611; rate2 = num/670 # rates of negative and positive samples
    if data.target[ind]:
        return np.random.choice([0, 1], size=(1,), p=[1- rate2, rate2])[0] == 1
    else:
        return np.random.choice([0, 1], size=(1,), p=[1- rate1, rate1])[0] == 1

def load_and_process_data(train_file, val_file, feature_size = None):
    data_train = pd.read_csv(train_file, sep = '\t').to_numpy()
    data_validation = pd.read_csv(val_file, sep = '\t').to_numpy()

    X_train = torch.tensor(data_train[:, 0:feature_size])
    X_validation = torch.tensor(data_validation[:, 0:feature_size])

    # normalize each input
    from sklearn.preprocessing import StandardScaler
    scaler_train = StandardScaler()
    scaler_train.fit(X_train)
    
    X_norm_train = scaler_train.transform(X_train)
    X_norm_validation = scaler_train.transform(X_validation)

    Y_train = torch.tensor(data_train[:, -1])
    Y_validation = torch.tensor(data_validation[:, -1])
    
    return X_norm_train, X_norm_validation, Y_train, Y_validation, scaler_train

def load_and_process_data_balanced(train_file, num = 500, feature_size = None):
    data = pd.read_csv(train_file, sep = "\t")
    ind = [x for x in data.index if select(data, x, num = num)] # indices to sample balanced classes
    data = data.loc[ind,].to_numpy()

    X = torch.tensor(data[:, 0:feature_size])

    # normalize each input
    from sklearn.preprocessing import StandardScaler
    scaler = StandardScaler()
    scaler.fit(X)
    
    X_norm = scaler.transform(X)
    Y = torch.tensor(data[:, -1])
    
    return X, X, Y, Y, scaler

# Create a train and test split. Use a seed for reproducability
def split_data(feature_vecs, Y):
    np.random.seed(0)
    num_data = len(Y)
    num_train = int(train_split * num_data)
    index = np.random.permutation(range(num_data))
    feat_vecs_train = feature_vecs[index[:num_train]]
    Y_train = Y[index[:num_train]]
    feat_vecs_test = feature_vecs[index[num_train:]]
    Y_test = Y[index[num_train:]]
    return feat_vecs_train, feat_vecs_test, Y_train, Y_test

def load_and_process_data_test(scaler, test_file, feature_size = None):
    data = pd.read_csv(test_file, sep = "\t")

    X = torch.tensor(data.iloc[:, 0:feature_size].to_numpy())
    print("First X sample, original  :", X[0])

    # normalize each input
    X_norm = scaler.transform(X)
    print("First X sample, normalized:", X_norm[0])
    Y = torch.tensor(data.target.to_numpy())

    return X_norm, Y
```

```python
def accuracy(labels, hard_predictions, weights):
    loss = 0
    for l, p in zip(labels, hard_predictions):
        if torch.abs(l - p) < 1e-5:
            loss += weights[int(l)]
    return loss

def print_statistic(cm, data_label):
    TP = cm.iloc[1,1]
    FP = cm.iloc[0,1]
    TN = cm.iloc[0,0]
    FN = cm.iloc[1,0]
    TPR = TP / (TP + FN)
    FPR = FP / (FP + TN)
    
    scaling = (TN + FP) / (TP + FN)
    TP_w = TP * scaling
    FN_w = FN * scaling
    PPV = TP / (TP + FP)
    NPV = TN / (TN + FN)
    F1  = 2*TP_w / (2*TP_w + FP + FN_w)
    
    print("TPR: {:0.7f} | FPR: {:0.7f} | TP: {} | FN: {} | TN: {} | FP: {}".format(TPR, FPR, TP, FN, TN, FP))
    print(data_label + " - TP_w: {:0.2f} | FN_w: {:0.2f} | PPV: {:0.7f} | NPV: {:0.7f} | F1: {:0.7f}".format(TP_w, FN_w, PPV, NPV, F1))

def training(features_train, features_validation, Y_train, Y_validation, iterations, lr = None, initial_parameters = None, num_classes = None, num_qubits = None, qnodes = None, num_layers = None, batch_size = None, margin = None):
    features = np.concatenate([features_train, features_validation], axis = 0)
    Y = torch.cat((Y_train, Y_validation), 0)
    
    num_data = Y.shape[0]
    num_train = Y.shape[0]
    q_circuits = qnodes

    # Initialize the parameters
    if initial_parameters is None:
        all_weights = [
            Variable(0.1 * torch.randn(num_layers, num_qubits, 3), requires_grad=True)
            for i in range(num_classes)
        ]
        all_bias = [Variable(0.1 * torch.ones(1), requires_grad=True) for i in range(num_classes)]
    else:
        all_weights = initial_parameters[0]
        all_bias = initial_parameters[1]
    optimizer = optim.Adam(all_weights + all_bias, lr=lr)
    params = (all_weights, all_bias)
    num_parameters = 0
    for entry in all_weights:
        num_parameters += np.prod(entry.shape)
    for entry in all_bias:
        num_parameters += np.prod(entry.shape)
    print("Num params: ", 3 * num_layers * num_qubits * 2 + 1 *2)
    print("Num params: ", num_parameters)

    costs, train_acc, val_acc = [], [], []
    
    # construct dictionary of weights
    label_count = torch.unique(Y, return_counts = True)
    labels = np.array(label_count[0])
    counts = np.array(label_count[1])
    print("All label counts:", counts)
    weights = 1 / (counts * len(counts))
    print("All label weights:", weights)
    weights = {**dict(zip(labels, weights))}
    
    label_count_train = torch.unique(Y_train, return_counts = True)
    labels_train = np.array(label_count_train[0])
    counts_train = np.array(label_count_train[1])
    print("Training label counts:", counts_train)
    weights_train = 1 / (counts_train * len(counts_train))
    print("Training label weights:", weights_train)
    weights_train = {**dict(zip(labels_train, weights_train))}
    
    label_count_validation = torch.unique(Y_validation, return_counts = True)
    labels_validation = np.array(label_count_validation[0])
    counts_validation = np.array(label_count_validation[1])
    print("Validation label counts:", counts_validation)
    weights_validation = 1 / (counts_validation * len(counts_validation))
    print("Validation label weights:", weights_validation)
    weights_validation = {**dict(zip(labels_validation, weights_validation))}

    # Train model and get performance on training and validation sets
    for it in range(iterations):
        print('starting iteration', it+1)
        
        optimizer.zero_grad() # set gradients to zero
        
        # Randomly sample batch to train on
        if (batch_size > 0):
            batch_index = np.random.randint(0, num_train, (batch_size,))
            feat_vecs_train_batch = features[batch_index]
            Y_train_batch = Y[batch_index]
            
            # Construct dictionary of weights for training batch
            label_count_train_batch = torch.unique(Y_train_batch, return_counts = True)
            labels_train_batch = np.array(label_count_train_batch[0])
            counts_train_batch = np.array(label_count_train_batch[1])

            weights_train_batch = 1 / (counts_train_batch * len(counts_train_batch))
            weights_train_batch = {**dict(zip(labels_train_batch, weights_train_batch))}
            
            curr_cost = multiclass_svm_loss(q_circuits, params, feat_vecs_train_batch, Y_train_batch, weights_train_batch, num_classes = num_classes, margin = margin)
        else:
            curr_cost = multiclass_svm_loss(q_circuits, params, features_train, Y_train, weights_train, num_classes = num_classes, margin = margin)
        
        curr_cost.backward()
        optimizer.step()

        # Compute predictions on train and validation sets
        predictions_train, score_train = classify(q_circuits, params, features_train, Y_train, num_classes = num_classes)
        predictions_validation, score_validation = classify(q_circuits, params, features_validation, Y_validation, num_classes = num_classes)
        predictions, score = classify(q_circuits, params, features, Y, num_classes = num_classes)
        
        # Compute accuracy on train and validation sets
        acc_train = accuracy(Y_train, predictions_train, weights_train)
        acc_validation = accuracy(Y_validation, predictions_validation, weights_validation)
        acc = accuracy(Y, predictions, weights)
        
        # Create confusion matrix
        cm_train = pd.crosstab(pd.Series(Y_train), pd.Series(predictions_train))
        cm_validation = pd.crosstab(pd.Series(Y_validation), pd.Series(predictions_validation))
        cm = pd.crosstab(pd.Series(Y), pd.Series(predictions))
        
        # Print intermediate statistics
        print("Iter: {:5d} | Cost: {:0.7f}".format(it + 1, curr_cost.item()))
        print("Training - Acc: {:0.7f} | ".format(acc_train), end = '')
        print_statistic(cm_train, "Training")
        print("Validation - Acc: {:0.7f} | ".format(acc_validation), end = '')
        print_statistic(cm_validation, "Validation")
        print("Whole - Acc: {:0.7f} | ".format(acc), end = '')
        print_statistic(cm, "Whole")

        costs.append(curr_cost.item())
        train_acc.append(acc_train)
        val_acc.append(acc_validation)

    return costs, train_acc, val_acc, q_circuits, params
```

```python
def training_wrapper(features_train, features_validation, Y_train, Y_validation, iterations, lr = None, initial_parameters=None, outputfile=None, outputfile_val=None, num_classes = None, num_qubits = None, qnodes = None, num_layers = None, batch_size = None, margin = None):
    t_0_sv = time.time()
    costs, train_acc, val_acc, q_circuits_original, params_original = training(features_train, features_validation, Y_train, Y_validation, iterations, lr, initial_parameters, num_classes = num_classes, qnodes = qnodes, num_layers = num_layers, num_qubits = num_qubits, batch_size = batch_size, margin = margin)
    t_1_sv = time.time()
    print("Execution time on sv device (seconds):", t_1_sv - t_0_sv)
    
    # Plot training cost and accuracy
    fig, ax1 = plt.subplots()
    iters = np.arange(0, len(train_acc), 1)
    colors = ["tab:red", "tab:blue"]
    ax1.set_xlabel("Iteration", fontsize=17)
    ax1.set_ylabel("Cost", fontsize=17, color=colors[0])
    ax1.plot(iters, costs, color=colors[0], linewidth=4)
    ax1.tick_params(axis="y", labelsize=14, labelcolor=colors[0])

    ax2 = ax1.twinx()
    ax2.set_ylabel("train Acc.", fontsize=17, color=colors[1])
    ax2.plot(iters, train_acc, color=colors[1], linewidth=4)

    ax2.tick_params(axis="x", labelsize=14)
    ax2.tick_params(axis="y", labelsize=14, labelcolor=colors[1])

    plt.grid(False)
    plt.tight_layout()
    if outputfile is not None:
        plt.savefig(outputfile) # save training plot
    
    # Plot validation cost and accuracy
    fig, ax1 = plt.subplots()
    iters = np.arange(0, len(val_acc), 1)
    colors = ["tab:red", "tab:blue"]
    ax1.set_xlabel("Iteration", fontsize=17)
    ax1.set_ylabel("Cost", fontsize=17, color=colors[0])
    ax1.plot(iters, costs, color=colors[0], linewidth=4)
    ax1.tick_params(axis="y", labelsize=14, labelcolor=colors[0])

    ax2 = ax1.twinx()
    ax2.set_ylabel("validation Acc.", fontsize=17, color=colors[1])
    ax2.plot(iters, val_acc, color=colors[1], linewidth=4)

    ax2.tick_params(axis="x", labelsize=14)
    ax2.tick_params(axis="y", labelsize=14, labelcolor=colors[1])

    plt.grid(False)
    plt.tight_layout()
    if outputfile_val is not None:
        plt.savefig(outputfile_val) # save validation plot


    return costs, train_acc, q_circuits_original, params_original
```

```python
def main(args):
    np.random.seed(0)
    torch.manual_seed(0)
    random.seed(10)
    np.random.seed(10)

    num_classes = 2
    margin= args.margin
    feature_size = 10
    batch_size = args.batch_size
    lr_adam = args.lr
    # the number of the required qubits is calculated from the number of features
    num_qubits = int(np.ceil(np.log2(feature_size)))
    num_layers = args.layer
    
    if args.device == "pennylane":
        dev = qml.device("default.qubit", wires=num_qubits)
    else:
        my_bucket = args.bucket
        my_prefix = args.folder
        s3_folder = (my_bucket, my_prefix)
        
# Uncomment when running on quantum computer        
#         if args.device == "SV1":
#             device_arn = "arn:aws:braket:::device/quantum-simulator/amazon/sv1"
#         elif args.device == "rigetti":
#             device_arn = "arn:aws:braket:::device/qpu/rigetti/Aspen-9"
#         elif args.device == "DM1":
#             device_arn = "arn:aws:braket:::device/quantum-simulator/amazon/dm1"
#         else:
#             sys.exit('device has to be pennylane, SV1, rigetti or DM1')

#         dev = qml.device(
#             "braket.aws.qubit",
#             device_arn=device_arn,
#             wires=num_qubits,
#             s3_destination_folder=s3_folder,
#             parallel=True,
#         )
    
    def layer(W):
        for i in range(num_qubits):
            qml.Rot(W[i, 0], W[i, 1], W[i, 2], wires=i)
        for j in range(num_qubits - 1):
            qml.CNOT(wires=[j, j + 1])
        if num_qubits >= 2:
            # Apply additional CNOT to entangle the last with the first qubit
            qml.CNOT(wires=[num_qubits - 1, 0])
        
    def circuit(weights, feat=None):
        qml.templates.embeddings.AmplitudeEmbedding(feat, range(num_qubits), pad=0.0, normalize=True)
        for W in weights:
            layer(W)

        return qml.expval(qml.PauliZ(0))

    qnodes = []
    for iq in range(num_classes):
        qnodes.append(qml.QNode(circuit, dev).to_torch())
    
    # Train from scratch or load model using pickle
    if args.trained_model is None:
        if args.balance :
            features_train, features_validation, Y_train, Y_validation, scaler = load_and_process_data_balanced(args.train_file, num = args.balance_size, feature_size = feature_size)
        else:
            features_train, features_validation, Y_train, Y_validation, scaler = load_and_process_data(args.train_file, args.val_file, feature_size = feature_size)
        
        if args.prefix is not None:
            outputfile = "%s.pdf"%(args.prefix)
            outputfile_val = "%s_val.pdf"%(args.prefix)
        else:
            outputfile = None
            outputfile_val = None
        costs, train_acc, q_circuits_original, params_original = training_wrapper(features_train, features_validation, Y_train, Y_validation, args.iteration, args.lr, initial_parameters = None, outputfile = outputfile, outputfile_val = outputfile_val, qnodes = qnodes, num_classes = num_classes, num_layers = num_layers, num_qubits = num_qubits, batch_size = batch_size, margin = margin)
        if args.prefix is not None:
            pickle.dump(params_original,open("%s.p"%args.prefix, "wb"))
            pickle.dump(scaler,open("%s_scaler.p"%args.prefix, "wb"))
        q_circuits_original[0].draw()
        
    else:
        params_original = pickle.load(open(args.trained_model, "rb"))
        scaler = pickle.load(open(args.trained_scaler, "rb"))
    
    # Evaluate model on test set and get accuracy
    if args.test_file is not None:
        features_test, Y_test = load_and_process_data_test(scaler, args.test_file, feature_size = feature_size)
        predictions_test, predScore = classify(qnodes, params_original, features_test, num_classes = num_classes)
        prediction = pd.concat([pd.DataFrame(predictions_test, columns = ["prediction"]),
                                pd.DataFrame(predScore, columns = ["0", "1"]),
                                pd.DataFrame(Y_test, columns = ["truth"])], axis = 1)
        if args.prefix is not None:
            prediction.to_csv("%s_prediction.tsv" % (args.prefix), sep = "\t", index = False)
        else:
            print(prediction)
        print("Accuracy: %s"%(str((prediction.truth == prediction.prediction).sum()/prediction.shape[0])))
        return prediction
```
